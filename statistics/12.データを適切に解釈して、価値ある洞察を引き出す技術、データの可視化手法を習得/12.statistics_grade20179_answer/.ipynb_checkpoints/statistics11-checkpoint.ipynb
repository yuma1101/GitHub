```python
import numpy as np
from collections import Counter

# 幹葉表示データを実データに変換
data = []

stem_leaf = {
    2: [0,0,1,3,3,5,6,8,9,9],
    3: [2,3,5,6,7,8,9,9,9],
    4: [0,1,3,3,4,4,4,4,5,9],
    5: [5,6,7,8,9,9],
    6: [4,6,7,7,9]
}

# 幹×10 + 葉 で実データに変換
for stem, leaves in stem_leaf.items():
    for leaf in leaves:
        data.append(stem * 10 + leaf)

data = np.array(data)
data.sort()

# 最頻値（mode）
freq = Counter(data)
mode = freq.most_common(1)[0][0]

# 中央値
median = np.median(data)

# 第一四分位数(Q1)
Q1 = np.percentile(data, 25)

print("データ数:", len(data))
print("最頻値:", mode)
print("中央値:", median)
print("第一四分位数:", Q1)
```

### ■ 問2　データの表示方法②

```python
# データ
total = 60
freq = {
    "0-20": 15,
    "21-40": 11,
    "41-60": int(0.27 * total),   # 16
    "61-80": 12,
    "81-100": 6
}

# 1. 41-60点の人数
num_41_60 = freq["41-60"]

# 2. 中央値が属する階級
import itertools

cumulative = list(itertools.accumulate(freq.values()))
median_pos1 = total // 2      # 30番目
median_pos2 = total // 2 + 1  # 31番目

classes = list(freq.keys())

def find_class(pos):
    for cls, cum in zip(classes, cumulative):
        if pos <= cum:
            return cls

median_class = find_class(median_pos1)

# 3. 平均点以上の人数（平均61 → 61〜80 と 81〜100）
above_mean = freq["61-80"] + freq["81-100"]

# 4. 65点が上位何％？
higher_than_65 = freq["81-100"]  # 65より上確定は81〜100
percentile = higher_than_65 / total * 100

print("41〜60点の人数:", num_41_60)
print("中央値が属する階級:", median_class)
print("平均点以上の人数:", above_mean)
print("65点のとき、少なくとも上位何%以内か:", percentile, "%以内")
```

### ■ 問3　ローレンツ曲線とジニ係数

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Mac標準の日本語フォントを指定（確実に入ってる）
plt.rcParams['font.family'] = 'Hiragino Sans'

# マイナス記号の文字化け対策
plt.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# データ定義
# -----------------------------
area = ["A", "B", "C", "D", "E"]
pop = np.array([40, 35, 30, 35, 10])     # 人口(万人)
tax = np.array([500, 1000, 1400, 2200, 2400])  # 税収(万円)

# -----------------------------
# 累積相対度数の計算
# -----------------------------
total_pop = pop.sum()
total_tax = tax.sum()

cum_pop = np.cumsum(pop) / total_pop
cum_tax = np.cumsum(tax) / total_tax

# 表としてまとめる
df = pd.DataFrame({
    "区画": area,
    "人口(万人)": pop,
    "税収(万円)": tax,
    "人口の累積相対度数": cum_pop,
    "税収の累積相対度数": cum_tax
})

print("▼ 累積相対度数つき表")
print(df)
print()

# -----------------------------
# ローレンツ曲線の描画
# -----------------------------
plt.plot(cum_pop, cum_tax, marker="o", label="Lorenz Curve")
plt.plot([0, 1], [0, 1], linestyle="--", label="Equality Line")
plt.xlabel("人口累積相対度数")
plt.ylabel("税収累積相対度数")
plt.title("Lorenz Curve")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# ジニ係数の計算（台形法）
# -----------------------------
lorenz_area = np.trapezoid(cum_tax, cum_pop)# ローレンツ曲線下の面積
#lorenz_area = np.trapz(cum_tax, cum_pop)   # ローレンツ曲線下の面積
gini = 1 - 2 * lorenz_area                 # ジニ係数

print("▼ ジニ係数:", gini)
```

### ■ 問4　指数化

```python
import pandas as pd
import numpy as np

# データ
data = pd.DataFrame({
    "year": [2020, 2021, 2022],
    "pencil_price": [70, 72, 75],      # 鉛筆の平均単価（円）
    "pencil_qty": [20, 22, 25],        # 鉛筆の販売数（千個）
    "eraser_price": [70, 75, 80],      # 消しゴムの平均単価
    "eraser_qty": [15, 20, 30]         # 消しゴムの販売数
})

# --------------------------------------------------
# 1️⃣ 鉛筆の平均単価指数（2020年＝100）
# --------------------------------------------------
base = data.loc[data['year'] == 2020, 'pencil_price'].iloc[0]
data["pencil_index"] = (data["pencil_price"] / base * 100).round(1)  # 小数点2位を四捨五入 → 1桁にしてもOK

# --------------------------------------------------
# 2️⃣ 単価変化率（％）
# --------------------------------------------------
rate_20_21 = (data.loc[1, "pencil_price"] - data.loc[0, "pencil_price"]) / data.loc[0, "pencil_price"] * 100
rate_21_22 = (data.loc[2, "pencil_price"] - data.loc[1, "pencil_price"]) / data.loc[1, "pencil_price"] * 100

bigger_rate = "2021→2022 の方が大きい" if rate_21_22 > rate_20_21 else "2020→2021 の方が大きい"

# --------------------------------------------------
# 3️⃣ ラスパイレス指数（基準年 = 2021 → 比較年 = 2022）
#   Laspeyres = Σ(p1*q0) / Σ(p0*q0)
# --------------------------------------------------
p0 = np.array([72, 75])     # 2021（鉛筆, 消しゴム）単価
p1 = np.array([75, 80])     # 2022 単価
q0 = np.array([22, 20])     # 2021 販売数

laspeyres = (p1 * q0).sum() / (p0 * q0).sum()
laspeyres = round(laspeyres, 4)

# --------------------------------------------------
# 4️⃣ パーシェ指数（基準年 = 2021 → 比較年 = 2022）
#   Paasche = Σ(p1*q1) / Σ(p0*q1)
# --------------------------------------------------
q1 = np.array([25, 30])     # 2022 販売数

paasche = (p1 * q1).sum() / (p0 * q1).sum()
paasche = round(paasche, 4)

# --------------------------------------------------
# 出力
# --------------------------------------------------
print("① 鉛筆の平均単価指数（2020年＝100）")
print(data[["year", "pencil_price", "pencil_index"]], "\n")

print("② 変化率:")
print(f"・2020→2021 : {rate_20_21:.2f}%")
print(f"・2021→2022 : {rate_21_22:.2f}%")
print(" → 大きいのは：", bigger_rate, "\n")

print("③ ラスパイレス指数（2021→2022）:", laspeyres)
print("④ パーシェ指数（2021→2022）:", paasche)

```

### ■ 問5　偏相関係数

```python
import seaborn as sns
import pandas as pd
import numpy as np

# データ読み込み
iris = sns.load_dataset("iris")

# x, y, z を指定
x = iris["sepal_length"]
y = iris["petal_width"]
z = iris["petal_length"]

# 1. 相関係数（xとy、yとz、xとz）
corr_xy = x.corr(y)
corr_yz = y.corr(z)
corr_xz = x.corr(z)

print("1. 相関係数")
print("x と y の相関係数:", corr_xy)
print("y と z の相関係数:", corr_yz)
print("x と z の相関係数:", corr_xz)

# 2. 偏相関係数（z の影響を除いた x と y）
# 偏相関係数の公式:
#   r_xy.z = (r_xy - r_xz * r_yz) / sqrt((1 - r_xz^2) * (1 - r_yz^2))

partial_corr_xy_z = (corr_xy - corr_xz * corr_yz) / np.sqrt((1 - corr_xz**2) * (1 - corr_yz**2))

print("\n2. z の影響を除いた x と y の偏相関係数:", partial_corr_xy_z)
```
