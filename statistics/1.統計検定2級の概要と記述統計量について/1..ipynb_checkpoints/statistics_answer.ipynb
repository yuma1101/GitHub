```python
import numpy as np
from scipy import stats
import pandas as pd 

# データ
data = [45, 40, 55, 60, 65, 95, 40, 30, 50, 55, 40]

# 最大値・最小値
max_value = np.max(data)
min_value = np.min(data)


# 平均値
mean_value = np.mean(data)

# 中央値
median_value = np.median(data)

# 最頻値
mode_result = stats.mode(data, keepdims=True)
mode_value = mode_result.mode[0]  # 最頻値
mode_count = mode_result.count[0] # 出現回数

# 第一四分位数・第三四分位数
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)


# 四分位範囲
iqr = q3 - q1


# # 結果表示
print(f"最大値: {max_value}")
print(f"最小値: {min_value}")
print(f"平均値: {mean_value}")
print(f"中央値: {median_value}")
print(f"最頻値: {mode_value} （{mode_count}回出現）")
print(f"第一四分位数: {q1}")
print(f"第三四分位数: {q3}")
print(f"四分位範囲: {iqr}")
```
