# 問2 Pythonを使って記述統計量の算出

```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
data = [45, 40, 55, 60, 65, 95, 40, 30, 50, 55, 40]
```

```python
df = pd.DataFrame(data)
```

## 1.最大値

```python
np.max(data)
```

```python
df.max()
```

## 2.最小値

```python
np.min(data)
```

```python
df.min()
```

## 3.平均値

```python
np.mean(data)
```

```python
df.mean()
```

## 4.中央値

```python
np.median(data)
```

```python
df.median()
```

## 5.最頻値

```python
stats.mode(data)
```

```python
df.mode()
```

## 6.第一四分位数

```python
np.percentile(data, 25)
```

```python
np.quantile(data, 0.25)
```

```python
df.quantile(0.25)
```

## 7.第三四分位数

```python
np.percentile(data, 75)
```

```python
np.quantile(data, 0.75)
```

```python
df.quantile(0.75)
```

2. 指定したパーセンテージの位置に値がない場合、近傍の2点で補完する（デフォルトでは線形補完）

## 8.四分位範囲

```python
q1, q3 = np.percentile(data, [25, 75])
```

```python
iqr = q3 - q1
iqr
```

```python
q1, q3 = np.quantile(data, [0.25, 0.75])
```

```python
iqr = q3 - q1
iqr
```

```python
df.quantile([0.25, 0.75])
```

# 問3 箱ひげ図の描画

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.boxplot(data)
plt.show()
```

```python
sns.boxplot(y=data)
plt.show()
```

外れ値・・・q3 + 1.5 * iqrより大きい値、またはq1 - 1.5 * iqrより小さい値<br>
q3 = 57.5<br>
q1 = 40<br>
iqr = 17.5<br>
1.5 * iqr = 26.25<br>
q3 + 1.5 * iqr = 83.75<br>
q1 - 1.5 * iqr = 13.75
