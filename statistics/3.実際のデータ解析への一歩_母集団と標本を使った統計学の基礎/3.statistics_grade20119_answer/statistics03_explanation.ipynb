# 150個の母集団の中から15個の標本を抽出する場合

## 1.単純無作為抽出法
完全にランダムに抽出する方法。

```python
sample1 = df.sample(15, random_state=1)
sample1
```

```python
sample1["species"].value_counts()
```

```python
sample1["petal_length"].mean()
```

```python
population_mean
```

乱数を変えて抽出

```python
sample1 = df.sample(15, random_state=100)
sample1
```

```python
sample1["species"].value_counts()
```

```python
sample1["petal_length"].mean()
```

```python
population_mean
```

メリット<br>
・最も単純でわかりやすい<br>
デメリット<br>
・均等に抽出されない場合もある<br>
・母集団によっては抽出が困難な場合もある

## 2.層化抽出法
母集団をあらかじめいくつかの部分母集団（層）に分けておき、各層の中から必要な数の標本を無作為に抽出する方法。<br>
【比例配分法】<br>
層化抽出法の中でも、母集団の各層の大きさに比例させて抽出数を配分し、抽出する方法。

<img src="01.png" width="70%">

例：母集団がオレンジ6割、青3割、緑1割なのでオレンジから6個、青から3個、緑から1個の合計10個を抽出する。

【ネイマン配分法】<br>
層化抽出法の中でも、母集団の各層の大きさと各層の標準偏差に比例させて抽出数を配分し、抽出する方法。<br>
バラつきが大きい層を優先して抽出する。

```python
df["species"].value_counts()
```

```python
df1 = df[df["species"] == "setosa"].sample(5, random_state=0)
df2 = df[df["species"] == "versicolor"].sample(5, random_state=0)
df3 = df[df["species"] == "virginica"].sample(5, random_state=0)
df_all = pd.concat([df1, df2, df3])
df_all
```

```python
df_all["petal_length"].mean()
```

```python
population_mean
```

メリット<br>
・推定精度が高い<br>
・各層で比較できる<br>
デメリット<br>
・母集団の構成情報を事前に知っておく必要がある

## 3.クラスター抽出法（集落抽出法）
母集団を小集団であるクラスター(集落)に分け、分けられたクラスターの中からいくつかのクラスターを無作為抽出する。<br>
それぞれのクラスターにおいて全数調査を行う方法。

<img src="02.png" width="70%">

例：母集団からクラスターを3つ作り、その中からランダムにクラスターを1つ選ぶ。選ばれたクラスター内のデータを全て調査する。

```python
df4 = df[df["species"] == "setosa"]
df4
```

```python
df4["petal_length"].mean()
```

```python
population_mean
```

メリット<br>
・標本の抽出が簡単<br>
デメリット<br>
・同じクラスターに属するデータは似た性質を持ちやすい。

https://bellcurve.jp/statistics/course/8007.html

## 4.系統抽出法
母集団に番号を振るなどして識別し、一定の規則に従って抽出する方法。

<img src="04.png" width="70%">

例：母集団に通し番号を付け、ランダムに数値を選び、選ばれた数値から6個ごとに調査対象を選ぶ。

```python
df5 = df.iloc[3::6]
df5
```

```python
df5["petal_length"].mean()
```

```python
population_mean
```

メリット<br>
・調査対象の選出が簡単<br>
デメリット<br>
・母集団の並び順に規則性があると誤差が大きくなりやすい

## 多段抽出法
母集団をいくつかのグループに分け、そこからランダムにいくつかグループを選ぶ。<br>
選ばれたグループをさらにいくつかのグループに分け、そこからランダムにいくつかグループを選ぶ。<br>
この操作を繰り返し、最終的に選ばれたグループの中からランダムに調査対象をいくつか選ぶ方法。<br>
【二段抽出法】<br>
多段抽出法の中でも、母集団からランダムにいくつかのグループを選び、そこからランダムに調査対象を選ぶという、二段に分けて抽出する方法。

<img src="03.png" width="70%">

例：母集団を5個のグループに分ける。そのグループの中から1個のグループをランダムに選ぶ。<br>
次に、選ばれたグループを3個のグループに分ける。そのグループの中から2個のグループをランダムに選ぶ。<br>
最終的に選ばれたグループを調査対象とする。

メリット<br>
・標本の抽出が簡単<br>
デメリット<br>
・グループに偏りがあると、誤差が大きくなりやすい

https://bellcurve.jp/statistics/course/8007.html

```python
from scipy import stats
```

```python
x = np.linspace(-5, 5, 101)
y = stats.norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.grid()
plt.vlines(x=1, ymin=0, ymax=0.25, color='orange', ls='--', linewidth=2)
plt.text(0.9, -0.02,"σ", fontsize=12)
plt.show()
```

σの値はグラフの変曲点のx座標

```python
x = np.linspace(-5, 5, 101)
y1 = stats.norm.pdf(x, loc=0, scale=1)
y2 = stats.norm.pdf(x, loc=1, scale=1)
y3 = stats.norm.pdf(x, loc=2, scale=1)
y4 = stats.norm.pdf(x, loc=-2, scale=1)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.grid()
plt.show()
```

y1 : $N(0, 1^{2})$<br>
y2 : $N(1, 1^{2})$<br>
y3 : $N(2, 1^{2})$<br>
y4 : $N(-2, 1^{2})$

# 問4 標準化

```python
x = np.linspace(-5, 10, 101)
y = stats.norm.pdf(x, loc=2, scale=2)
```

```python
plt.plot(x, y)
plt.vlines(x=2, ymin=0, ymax=0.2, color='gray', ls='--', linewidth=2)
plt.text(2, -0.001,"μ", fontsize=12)
plt.vlines(x=4, ymin=0, ymax=0.125, color='gray', ls='--', linewidth=2)
plt.hlines(xmin=2, xmax=4, y=0.075, color='gray', ls='--', linewidth=2)
plt.text(2.9, 0.081,"σ", fontsize=12)
plt.grid()
plt.show()
```

$\mu=2$, $\sigma=2$

1.平均値を引く

```python
x1 = x - 2
```

```python
plt.plot(x, y, ls="--")
plt.plot(x1, y)
plt.vlines(x=0, ymin=0, ymax=0.2, color='gray', ls='--', linewidth=2)
plt.text(0, -0.001,"0", fontsize=12)
plt.vlines(x=2, ymin=0, ymax=0.125, color='gray', ls='--', linewidth=2)
plt.hlines(xmin=0, xmax=2, y=0.075, color='gray', ls='--', linewidth=2)
plt.text(0.9, 0.081,"σ", fontsize=12)
plt.grid()
plt.show()
```

$\mu=0$, $\sigma=2$

2.標準偏差で割る

```python
x2 = x1 / 2
```

```python
plt.plot(x1, y, ls="--")
plt.plot(x2, y)
plt.vlines(x=0, ymin=0, ymax=0.2, color='gray', ls='--', linewidth=2)
plt.text(0, -0.001,"0", fontsize=12)
plt.vlines(x=1, ymin=0, ymax=0.125, color='gray', ls='--', linewidth=2)
plt.hlines(xmin=0, xmax=1, y=0.075, color='gray', ls='--', linewidth=2)
plt.text(0.3, 0.081,"1", fontsize=12)
plt.grid()
plt.show()
```

$\mu=0$, $\sigma=1$

データが$N(\mu, \sigma^{2})$に従うとき、
$$u = \frac{x-\mu}{\sigma}$$
は$N(0, 1^{2})$に従う。
