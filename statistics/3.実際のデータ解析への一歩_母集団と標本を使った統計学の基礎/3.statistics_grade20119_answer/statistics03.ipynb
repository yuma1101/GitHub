```python
from scipy import stats
x = np.linspace(-5,5,101)
y = stats.norm.pdf(x,loc=0, scale=1)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#σの値は、平均値からグラフの変曲点までの距離
plt.plot(x,y)
plt.grid()
plt.vlines(x=1,ymin=0, ymax=0.25, color='orange', ls = '--', linewidth=2)
plt.text(0.9,-0.02, "σ", fontsize=12)
plt.show()
```

```python
x = np.linspace(-5, 5,101)
y1 = stats.norm.pdf(x,loc = 0, scale =1)
y2 = stats.norm.pdf(x,loc = 1, scale =1)
y3 = stats.norm.pdf(x,loc = 2, scale =1)
y4 = stats.norm.pdf(x,loc = -2, scale =1)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid()
plt.show()
```
