# 問2 Pythonを使って適合度検定

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
obs = [44, 32, 68, 54, 46, 56]
exp = [50, 50, 50, 50, 50, 50]
```

```python
chi2, p = stats.chisquare(obs, exp)
chi2, p
```

p値 < 有意水準0.025(両側で)　なので帰無仮説H0を棄却する

グラフに描画

```python
dof = len(obs) - 1
alpha = 0.05
```

```python
x = np.linspace(0, 20, 101)
y = stats.chi2.pdf(x, dof)
```

```python
lower_cv = stats.chi2.ppf(alpha/2, dof)
upper_cv = stats.chi2.ppf(1-alpha/2, dof)
(lower_cv, upper_cv)
```

```python
plt.plot(x, y)
plt.vlines(x=chi2, ymin=0, ymax=0.15, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(0, lower_cv, 100)
plt.fill_between(lower_x, stats.chi2.pdf(lower_x, dof), facecolor='orange')

upper_x = np.linspace(upper_cv, 20, 100)
plt.fill_between(upper_x, stats.chi2.pdf(upper_x, dof), facecolor='orange')
plt.grid()
plt.show()
```

# 問4 Pythonを使って独立性の検定

```python
obs = np.array([[120, 24], 
                [240, 32], 
                [150, 10], 
                [190, 34]])
```

```python
chi2, p, dof, exp = stats.chi2_contingency(obs)
chi2, p
```

```python
exp
```

p値 < 有意水準0.025(両側で)　なので帰無仮説H0を棄却する

グラフに描画

```python
dof = (obs.shape[0] - 1) * (obs.shape[1] - 1)
alpha = 0.05
```

```python
x = np.linspace(0, 20, 101)
y = stats.chi2.pdf(x, dof)
```

```python
lower_cv = stats.chi2.ppf(alpha/2, dof)
upper_cv = stats.chi2.ppf(1-alpha/2, dof)
(lower_cv, upper_cv)
```

```python
plt.plot(x, y)
plt.vlines(x=chi2, ymin=0, ymax=0.15, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(0, lower_cv, 100)
plt.fill_between(lower_x, stats.chi2.pdf(lower_x, dof), facecolor='orange')

upper_x = np.linspace(upper_cv, 20, 100)
plt.fill_between(upper_x, stats.chi2.pdf(upper_x, dof), facecolor='orange')
plt.grid()
plt.show()
```
