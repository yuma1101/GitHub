# 問3 Pythonを使って不偏分散の算出

```python
import numpy as np
import pandas as pd
```

```python
data = [8.0, 5.4, 3.6, 5.8, 6.4, 6.2]
```

## 1.標本分散

```python
np.var(data)
```

## 2.不偏分散

```python
np.var(data, ddof=1)
```

<img src="02.png" width=100%>

分散の自由度はN - ddofで表される。<br>
デフォルトではddofには0が渡されているので、デフォルトでは標本分散が算出される。<br>
ddofに1を渡すと、自由度がN - 1になるので、不偏分散が算出される。

# 問5 Pythonを使って母平均の区間推定（母分散既知）

```python
from scipy import stats
```

```python
data2 = [10.3, 10.5, 9.8, 10.2, 10.4, 10.8, 10.7, 10.7, 9.7, 9.9]
```

## 1.点推定

```python
np.mean(data2)
```

## 2.区間推定

```python
ci = 0.95 #信頼区間(confidence interval)
```

```python
barx = np.mean(data2) #標本平均
```

```python
se = np.sqrt(0.4**2 / len(data2)) #標準誤差(統計量の標準偏差, standard error)
```

```python
lower, upper = stats.norm.interval(ci, barx, se)
```

```python
print(f'下限={lower:.2f}\n上限={upper:.2f}')
```

# 問7 Pythonを使って母平均の区間推定（母分散未知）

```python
data2 = [10.3, 10.5, 9.8, 10.2, 10.4, 10.8, 10.7, 10.7, 9.7, 9.9]
```

## 1.点推定

```python
np.mean(data2)
```

## 2.区間推定

```python
ci = 0.95 #信頼区間(confidence interval)
```

```python
barx = np.mean(data2) #標本平均
s = np.var(data2, ddof=1) #不偏分散 stats.tvar(data2)でも可
dof = len(data2) - 1 #自由度(degree of freedom)
```

```python
se = np.sqrt(s / len(data2)) #標準誤差(統計量の標準偏差, standard error)
```

```python
lower, upper = stats.t.interval(ci, dof, barx, se)
```

```python
print(f'下限={lower:.2f}\n上限={upper:.2f}')
```

# 問9 Pythonを使って母分散の区間推定

```python
data3 = [10, 9, 9, 9, 10, 10, 11, 9, 9, 9]
```

## 1.点推定

```python
np.var(data3, ddof=1)
```

## 2.区間推定

```python
ci = 0.95 #信頼区間(confidence interval)
```

```python
s = np.var(data3, ddof=1) #不偏分散 stats.tvar(data3)でも可
dof = len(data3) - 1 #自由度(degree of freedom)
S = s * dof #平方和
```

```python
chi2_0975, chi2_0025 = stats.chi2.interval(ci, dof)
```

```python
lower = S / chi2_0025
upper = S / chi2_0975
```

```python
print(f'下限={lower:.2f}\n上限={upper:.2f}')
```
