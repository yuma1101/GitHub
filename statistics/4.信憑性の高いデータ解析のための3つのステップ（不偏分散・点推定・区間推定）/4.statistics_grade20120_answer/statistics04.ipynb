```python
#問5　母平均の推定
from scipy import stats

data2 = [10.3, 10.5, 9.8, 10.2, 10.4, 10.8, 10.7, 10.7, 9.7, 9.9]
```

## 1.点推定

```python
np.mean(data2)
```

## 2.区間推定

```python
ci = 0.95#信頼区間(confidence interval)
```

```python
barx = np.mean(data2)# 標本平均

se = np.sqrt(0.4 ** 2 / len(data2)) #標準誤差(統計量の標準偏差、standard error)
```

```python
lower, upper = stats.norm.interval(ci, barx,se)
```

```python
print(f'下限={lower:.2f}\n上限={upper:.2f}')
```

```python
#問7 pythonを用いて母平均の区間推定(母分散未知)

data2 = [10.3, 10.5, 9.8, 10.2, 10.4, 10.8, 10.7, 10.7, 9.7, 9.9]
```

## 1.点推定

```python
np.mean(data2)
```

## 2.区間推定

```python
ci = 0.95 # 信頼区間(confidence interbal)
```

```python
barx = np.mean(data2)
s = np.var(data2,ddof=1)
dof = len(data2) - 1
```

```python
se = np.sqrt(s / len(data2))
```

```python
lower2,upper2 = stats.t.interval(ci, dof, barx, se)
```

```python
print(f'下限={lower2:.2f}\n上限={upper2:.2f}')
```

```python
#問9 pythonを用いて母分散の区間推定

data3 = [10,9,9,9,10,10,11,9,9,9]
data3
```

## 1.点推定

```python
np.var(data3,ddof=1)
```

## 2.区間推定

```python
ci = 0.95 #信頼区間(confidence interval)
```

```python
s = np.var(data3,ddof=1)
dof = len(data3) - 1#自由度(degree of freedom)
S = s * dof #平方和
```

```python
chi2_0975,chi2_0025 = stats.chi2.interval(ci,dof)
```

```python
lower = S / chi2_0025
upper = S / chi2_0975
```

```python
print(f"下限={lower:.2f}\n上限={upper:.2f}")
```
