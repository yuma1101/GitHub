```python
%%bigquery --project propane-zephyr-478723-n2
SELECT 
    is_male, 
    COUNT(*) as count,
    AVG(weight_pounds) as avg_weight
FROM `bigquery-public-data.samples.natality`
GROUP BY is_male;
```

```python
import pandas as pd
from google.cloud import bigquery

# 1. CSVを読み込む
df = pd.read_csv('your_data.csv')

# 2. BigQueryクライアントの初期化
client = bigquery.Client(project='あなたのプロジェクトID')

# 3. 格納先のテーブル指定（データセット名.テーブル名）
table_id = "your_dataset.your_table_name"

# 4. データを送信（テーブルがなければ自動作成される）
job = client.load_table_from_dataframe(df, table_id)
job.result()  # 完了を待つ

print(f"Loaded {df.shape[0]} rows to {table_id}")
```
